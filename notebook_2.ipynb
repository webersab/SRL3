```python
import spacy
from spacy import displacy
from allennlp.predictors.predictor import Predictor
import re
import utils
```

```python
def find_arguments(input_string):
    # Define the regex pattern for ARG1, ARG2, ARG3, and ARG4
    pattern = r'\[ARG[1-4]: (.*?)\]'
    
    # Find all matches in the input string
    matches = re.findall(pattern, input_string)
    
    return matches

print(find_arguments("this is a test [ARG1: yolo]"))
```

```python
SRL_MODEL_PATH = "srl.tar.gz"
COREF_MODEL_PATH = "coref.tar.gz"
```

```python
predictor_srl = Predictor.from_path(SRL_MODEL_PATH)
predictor_coref = Predictor.from_path(COREF_MODEL_PATH)
nlp = spacy.load("en_core_web_sm")
```

```python
with open('a_little_princess.txt', 'r') as file:
    content = file.read(10000)
print("file read")

DOCUMENT = content
#DOCUMENT = "Chris Mitchell left the house and went to the garden. He met Melissa Smith and proceeded to climb the wall. Next the man jumped into the rose bush and she watched him."
```

```python
predictions_coref = predictor_coref.predict(document=DOCUMENT)
print(predictions_coref)
```

```python
doc = nlp(DOCUMENT)

for ent in doc.ents:
    print(f"Entity: {ent.text}, Type: {ent.label_}, Start Token: {ent.start}, End Token: {ent.end}")
```

```python
protagonist_dict = {}

for idx, i in enumerate(predictions_coref["clusters"]):
    for ent in doc.ents:
        if utils.contains_number(ent.start,i):
             for sublist in i:
                for number in sublist:
                # Assign the specific string to the number in the dictionary
                    protagonist_dict[number] = ent.text
print(protagonist_dict)
```

```python
# Input string
text = DOCUMENT

# Process the text with spaCy
doc = nlp(text)

# Break into sentences and create mapping of global indices
sentence_info = []

global_token_idx = 0
sentence_index = 1
word_indices_dict = {}
sentences = []

for sent in doc.sents:
    sentences.append(sent.text)
    word_indices_list = []
    
    # Get indices of words in the global string
    
    for token in sent:
        global_index = global_token_idx  # Start index of the token in the original string
        word_indices_list.append(global_index)  # Store the index and word text
        global_token_idx += 1

    word_indices_dict[sentence_index] = word_indices_list
    sentence_index += 1

print(word_indices_dict)
```

```python
sentence_counter = 1
for sentence in sentences:
    #print("----------")
    #print('sent',sentence_counter)
    predictions_srl = predictor_srl.predict(sentence=sentence)
    #print(predictions_srl)

    for n in predictions_srl['verbs']:
        global_index = -1
        arg0_index = -1
        arg1_index = -1
        
        if "B-ARG0" in n['tags']:
            arg0_index = n['tags'].index("B-ARG0")
            global_index_arg0 = word_indices_dict[sentence_counter][arg0_index]
        else:
            break
        if "B-ARG1" in n["tags"]:
            arg1_index = n['tags'].index("B-ARG1")
            global_index_arg1 = word_indices_dict[sentence_counter][arg1_index]

        protagonist = 0

        if global_index_arg0 in protagonist_dict.keys():
            protagonist = protagonist_dict[global_index_arg0]
            if global_index_arg1 in protagonist_dict.keys():
                antagonist = protagonist_dict[global_index_arg1]
                print(protagonist, utils.lemmatize(n['verb'], nlp), antagonist)
            else:
                arguments = find_arguments(n['description'])
                print(protagonist, utils.lemmatize(n['verb'], nlp), arguments)
    sentence_counter += 1


```
